```python
# renaming two columns for my own clarity
# I wanted to change the message column into a str data type, but none of the rows had any value in them

df.rename(columns={'name': 'indicator_name', 'message': 'message_or_notes'}, inplace=True)
```

```python
# it's probably better if I just get rid of the message column aka message_or_notes column 
# since no observations have values in that column

df.drop(columns='message_or_notes', inplace=True)

df
```

```python
# dropped the first row here, rendering the rest of the df

df.drop(index=0, inplace=True)

df
```

```python
# sortingt two columns, both ascending, which is the default setting
df.sort_values(['geo_place_name', 'data_value'], inplace=True)

df
```

```python
# Filtering based on geo_join_id where the id is 503
willowbrook = df[df.geo_join_id == 503]
```

```python
# This code triggers a warning, but I believe it is fine since I did intend to subset the original dataframe
# This converts the geo_place_name column values into lowercase string

willowbrook['geo_place_name'] = willowbrook['geo_place_name'].apply(str.lower)

willowbrook
```

```python
# Grouping by geo_join_id column and getting the mean of data_value from each geo_join_id
mean_geo_join_id = df.groupby("geo_join_id")["data_value"].mean()


mean_geo_join_id
```

```python
df.groupby("geo_join_id")["data_value"].min()
```

```python
df.groupby("geo_join_id")["data_value"].max()
```

```python
# Here I am grouping by measure first, and then geo_place_name and getting the max value by the grouped criteria
grouped = df.groupby(['measure', 'geo_place_name'])['data_value'].max()

# This kind of surprised me because the output is now a Series and not another dataframe
grouped.sort_values()
```
